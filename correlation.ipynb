# Correlation Matrix

```python
import pandas as pd
steam_ds = pd.read_csv('steam_games_cleaned.csv')
steam_ds
```

```python
numeric_df = steam_ds[['hours_played', 'recommendation']]
corr_matrix = numeric_df.corr()
corr_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

## Summary of Analysis

- The matrix compares two key numeric variables: hours_played and recommendation.
- It quantifies the linear relationship between playtime and sentiment

## Trends and Relationships

- The correlation between hours_played and recommendation is 0.01, indicating a very weak positive relationship.
- This confirms what your boxplot suggested: while players who spend more time may be slightly more likely to recommend a game, the effect is minimal.
- The diagonal values (1.00) show perfect self-correlation, as expected.

## Insights

- The weak correlation suggests that playtime alone doesn’t strongly predict sentiment — players may recommend or reject a game regardless of how long they played.
- This highlights the importance of other factors (game quality, expectations, genre fit) in shaping user satisfaction.
- It also reinforces the value of combining multiple metrics (playtime, session frequency, retention) to understand engagement more holistically.

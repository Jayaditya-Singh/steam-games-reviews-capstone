# Steam GigaSheet Preprocessing and Analysis

```python
import pandas as pd
steam_ds = pd.read_csv("Steam Game Reviews export 2025-12-14 23-26-22.csv")
steam_ds
```

```python
steam_ds.isnull().sum()
```

```python
steam_ds = steam_ds.dropna()
steam_ds.isnull().sum()
```

```python
steam_ds
```

```python
steam_ds = steam_ds.drop_duplicates()
steam_ds
```

```python
steam_ds['hours_played'] = pd.to_numeric(steam_ds['hours_played'], errors='coerce')
steam_ds['recommendation'] = steam_ds['recommendation'].map({'Recommended':1, 'Not Recommended':0})

steam_ds
```

```python
def fix_date(val):
    try:
        return pd.to_datetime(val)
    except:
        return pd.to_datetime(val + '/2025', errors='coerce')
steam_ds['date'] = steam_ds['date'].astype(str).apply(fix_date)
```

```python
steam_ds
```

```python
steam_ds['review'] = steam_ds['review'].str.lower().str.replace('[^\w\s]', '', regex=True)
steam_ds
```

```python
steam_ds['playtime'] = pd.cut(steam_ds['hours_played'], bins=[0,10,100,1000], labels=['Casual', 'Moderate', 'Heavy'])
steam_ds
```

```python
steam_ds.to_csv("steam_games_cleaned.csv", index=False)
```
